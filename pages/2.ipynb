```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Meiryo"
```

```python
df = pd.read_csv("data/上場企業.csv")
df
```

```python
df.isnull().sum()
```

```python
df.groupby(["17業種区分"]).agg("count")
```

```python
#上場企業の業種分布はどうなっているか調べる
class_count = df["17業種区分"].value_counts()
type(class_count)
```

```python
plt.bar(class_count.index, class_count)
plt.show()
```

```python
plt.barh(class_count.index, class_count)
plt.show()
```

```python
class_count.sort_values(inplace=True)
plt.barh(class_count.index, class_count)
plt.show()
```

```python
c_list = ["gray"] * len(class_count)
c_list[-1] = "red"
plt.barh(class_count.index, class_count, color=c_list)
plt.show()
```
